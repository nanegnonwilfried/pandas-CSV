```python
# calculer la médiane et la moyenne des colonnes  Age et Customer spendings
# la mediane de la variable Age et de la variable Customer spendings
data['Age'].median()
data['Customer spendings'].median()
```

```python
# la moyenne de la variable Age et de la variable Customer spendings
data['Age'].mean()
data['Customer spendings'].mean()
```

```python
# recoder l’age en variable qualitative ne faisant des inter
def inter20(w):
  if w < 38:
    return "JEUNE"
  elif 38<=w <58:  
    return "ADULT"
  elif 58<= w:
    return "VIEUX"
    
```

```python
data['Age'].apply(inter20)
```

```python
# Calculer la médiane d’âge pour chaque pays
data.groupby(['Country'])['Age'].median()
```

```python
# Créer une visualisation des données du graphique à barres qui montre les dépenses des clients par pays
data.groupby(['Country'])['Customer spendings'].sum().plot(kind="bar",figsize=(15,6))

```

```python
df=data.copy()
df
```

```python
# Nettoyer les lignes avec moins de 10 € de dépenses par client : supprimez toutes les lignes d’utilisateurs ayant dépensé moins de 10 € sur le site,
df= df.drop(df[df['Customer spendings']<10].index)
df
```

```python
# Nettoyer les doublons : supprimer toutes les lignes qui apparaissent plus d’une fois dans les données
df=df.drop_duplicates()
df
```

```python
#  Écrire les données nettoyées dans un nouveau fichier CSV avec uniquement les colonnes suivantes 
df1=df[['Country','Age','Gender','Customer spendings']]
df1.to_csv("project1.csv")
```
